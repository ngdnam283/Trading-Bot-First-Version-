```python
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
api_key = os.getenv('BINANCE_API_KEY')
api_secret = os.getenv('BINANCE_SECRET_KEY')
```

```python
from binance.client import Client 
```

```python
client = Client(api_key, api_secret, testnet=True)
```

```python
account_info = client.get_account()

# Loop through the balances and sell each asset with a non-zero balance
for balance in account_info['balances']:
    asset = balance['asset']
    free_balance = float(balance['free'])

    # Skip if there's no balance to sell
    if free_balance > 0:
        # Example: Selling the asset for USDT (you can adjust the symbol as needed)
        symbol = f'{asset}USDT'
        try:
            # Place a market sell order for the entire available balance
            order = client.order_market_sell(
                symbol=symbol,
                quantity=free_balance
            )
            print(f"Sold {free_balance} {asset} for USDT")
        except Exception as e:
            print(f"Error selling {asset}: {e}")
```
